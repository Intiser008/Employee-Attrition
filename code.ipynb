```python
#Importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importing dataset
#Link to dataaset: https://www.kaggle.com/datasets/pavansubhasht/ibm-hr-analytics-attrition-dataset?resource=download
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head(10)
```

```python
#Checking the data

df.describe()
```

```python
#Visualizing the employee attrition
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Attrition'] = le.fit_transform(df['Attrition'])
sns.histplot(df['Attrition'], discrete=True)
df['Attrition'].value_counts() # 0 = False, 1 = True
```

```python
df
```

```python
#Show the number of employees that left and stayed by their age
plt.subplots(figsize = (12,4))
sns.countplot(x = 'Age', hue='Attrition', data = df, palette = 'colorblind')

#Conclusion: People at the age group of 34-36 tend to leave the company the most
```

```python
#Print all the data types of the attributes and their unique values
for col in df.columns:
    if df[col].dtype == object:
        print(str(col) + " : " + str(df[col].unique()))
        print(df[col].value_counts())
        print("___________________________________________________________________________________")
```

```python
#Data cleaning

df = df.drop('Over18', axis = 1) #all are 'Y'
df = df.drop('EmployeeNumber', axis = 1) #unique for each employee, so no correlation to attrition
df = df.drop('StandardHours', axis = 1) #only has 80 unique values
df = df.drop('EmployeeCount', axis = 1)
```

```python
df
```

```python
#Correlation between the remaining columns
plt.figure(figsize=(18,18))
sns.heatmap(df.corr(), annot = True, fmt = '.0%')
```

```python
#Transforming non-numerical columns to numerical columns

for col in df.columns:
    if df[col].dtype != np.number:
        df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
#Bringing the attrition column in front
df['Age_Years'] = df['Age']
df = df.drop('Age', axis = 1)
df
```

```python
X = df.drop('Attrition', axis = 1)
Y = df['Attrition']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

#Train:Test ratio is 75:25
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42)
```

```python
#Using the random forest classifier
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
forest.fit(X_train, Y_train)
```

```python
#Get the accuracy on the training dataset
forest.score(X_train, Y_train)
```

```python
#Get the accuracy on test data
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(Y_test, forest.predict(X_test))

TN = cm[0,0]
TP = cm[1,1]
FN = cm[1,0]
FP = cm[0,1]

ConfusionMatrixDisplay(cm).plot()

print('Model Teesting Accuracy = {}'.format((TP + TN)/(TP+TN+FN+FP)))
```
